```python
import pandas as pd
import sqlite3
```

```python
rt_reviews_df = pd.read_csv('zippedData/rt.reviews.tsv.gz',delimiter='\t',encoding='windows-1252')
rt_reviews_df
```

```python
rt_df = pd.read_csv('zippedData/rt.movie_info.tsv.gz',delimiter='\t',encoding='windows-1252')
rt_df
```

```python
rt_movies_df = rt_reviews_df.merge(rt_df, on = "id", how = "left")
rt_movies_df
```

```python
rt_movies_df['genre'] = rt_movies_df['genre'].str.split("|")
rt_movies_df
```

```python
rt_movies_df = rt_movies_df.explode("genre")
rt_movies_df
```
