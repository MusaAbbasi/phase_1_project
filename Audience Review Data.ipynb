```python
imdb_df = pd.read_sql("""
SELECT *
FROM movie_basics AS tb1
LEFT JOIN movie_ratings AS tb2
    ON tb1.movie_id = tb2.movie_id
""",con)
imdb_df
```

```python
imdb_df = imdb_df.dropna()
imdb_df
```

```python
imdb_df.groupby("genres").count()
```

```python
imdb_df = imdb_df.loc[:,~imdb_df.apply(lambda x: x.duplicated(),axis=1).all()].copy() #removed duplicate columns
imdb_df
```

```python
imdb_df['genres'] = imdb_df['genres'].str.split(",")
imdb_df['genres']
```

```python
imdb_df = imdb_df.explode("genres")
```
